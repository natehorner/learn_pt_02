```python
#common
import torch
from torch import nn
SEED_BASE = int("BADBEEF",16)
device = "cuda" if torch.cuda.is_available() else "cpu"

if(device == "cpu"):
  torch.manual_seed(SEED_BASE)
else:torch.cuda.manual_seed(SEED_BASE)

#
import matplotlib.pyplot as plt

xvals = torch.arange(-10,10,.1,dtype=torch.float32)

```

```python
#ReLU function: y=max(0,x)
plt.scatter(x=xvals,y=torch.relu(xvals))
```

```python
#sigmoid y = 1/(1+e^-x)
plt.scatter(x=xvals,y=torch.sigmoid(xvals))
```

```python
#sinc function y=sin(x)/x
plt.scatter(x=xvals,y=torch.sinc(xvals))
#a good wavelet that is semi-finite in both frequency and time
```

```python
#FFT of Sinc function why not - doesn't look square
#I think I learned it at one point but its gone now.
from torch import fft
plt.scatter(x=xvals,y=torch.abs(torch.fft.fft(torch.sinc(xvals))))
```
